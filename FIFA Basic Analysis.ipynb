```python
#FIFA DATASET
import pandas as pd 
fifa=pd.read_csv('fifa.csv')
from matplotlib import pyplot as plt 
import seaborn as sns 
import numpy as np 
fifa.shape
```

```python
list(fifa['nationality'].value_counts()[0:5].keys())
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color=['red','green','yellow','blue','pink'])
plt.show()
```

```python
player_salary= fifa[['short_name','wage_eur']]
player_salary[0:5]
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(7.5,7))
plt.bar(list(player_salary['short_name'][0:5]),list(player_salary['wage_eur'][0:5]),color=['red','yellow','blue','green','purple'])
plt.show()
```
