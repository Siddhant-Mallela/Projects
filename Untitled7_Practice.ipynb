```python
def f(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        val=row['Home Team Name'],'won the match'
    elif row['Home Team Goals'] < row['Away Team Goals']:
        val=row['Away Team Name'],'won the match'
    else:
        val='Match Draw'
    return val

matches['Result'] = matches.apply(f, axis=1)
matches.dtypes
```

```python
matches.head(5)
```

```python
def g(row):
    if row['Home Team Goals']>row['Away Team Goals']:
        val=row['Home Team Name']
    elif row['Home Team Goals']<row['Away Team Goals']:
        val=row['Away Team Name']
    else:
        val='-'
    return val
        
matches['winner']=matches.apply(g,axis=1)
matches.head(4)
```

```python
#Print Score of football matches in that order

def r(row):
    if row['Home Team Goals']>=row['Away Team Goals']:
        val=row['Home Team Goals'],'-',row['Away Team Goals']
    else:
        val=row['Home Team Goals'],'-',row['Away Team Goals']
    return val
matches['Score']=matches.apply(r,axis=1)
matches.head(5)
```

#### HOW MANY OF THEM BOUGHT 2,3,4 SAME ITEMS?

```python
from itertools import combinations
from collections import Counter

count= Counter()
for row in dupli['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
                 
print(count)
```

```python
count.most_common(10)
```

```python
for keys,values in count.most_common(10):
    print(keys,values)
```
